```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('VideoAndFrames/1.jpg')

#cv2.imshow('Original Image', img)
plt.imshow(img)
plt.show()

# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img, (3,3), sigmaX=0, sigmaY=0)
# cv2.imshow('Blurred Image', img_blur)
```

```python
# Sobel Edge Detection
from cProfile import label


sobel_x = cv2.Sobel(img,cv2.CV_64F, 1, 0, ksize=5) # According to the x direction
sobel_y = cv2.Sobel(img,cv2.CV_64F, 0, 1, ksize=5) # According to the y direction
sobel_xy = cv2.Sobel(img,cv2.CV_64F, 1, 1, ksize=5) # According to the x and y direction
sobel_y_blurr = cv2.Sobel(img_blur,cv2.CV_64F, 0, 1, ksize=5)

#cv2.imshow('Sobel X', sobel_x)
#cv2.imshow('Sobel Y', sobel_y)
#cv2.imshow('Sobel XY', sobel_xy)
#cv2.imshow('Sobel Y Blurred', sobel_y_blurr)
#cv2.waitKey(0)

plt.imshow(sobel_y)
plt.show()
plt.imshow(sobel_y_blurr)
plt.show()
```

## Results

With a blurred image and a Sobek detection we can clearly see the interface.
